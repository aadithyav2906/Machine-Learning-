```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
from sklearn.preprocessing import StandardScaler
from sklearn.model_selection import train_test_split
from sklearn import svm
from sklearn.svm import SVC
from sklearn.metrics import accuracy_score
```

```python
diabetes_data=pd.read_csv("/content/diabetes (1).csv")
diabetes_data.head()
```

```python
diabetes_data.describe()
```

```python
diabetes_data.groupby("Outcome").mean()
```

```python
diabetes_data["Outcome"].value_counts()
```

```python
x=diabetes_data.drop(columns="Outcome",axis=0)
y=diabetes_data["Outcome"]
```

Standardising the data

```python
scaler=StandardScaler()
x=scaler.fit_transform(x)
print(x)
```

```python
x_train,x_test,y_train,y_test=train_test_split(x,y,test_size=0.2,random_state=5)
model=SVC(kernel="linear")
model.fit(x_train,y_train)
#predicting the train data
y_train_predict=model.predict(x_train)
print(accuracy_score(y_train,y_train_predict))
#predicting the test data
y_test_predict=model.predict(x_test)
print(accuracy_score(y_test,y_test_predict))
```

```python
data=(7,107,74,0,0,29.6,0.254,31)
array1=np.array(data)
array2=array1.reshape(1,-1)
prediction=model.predict(array2)
if prediction[0]==1:
  print("He/She is diabetic")
else:
  print("He/She is healthy")
```
